# The Fastest List in the West

Create a list of 1 million numbers using Python lists.

Create a NumPy array of 1 million numbers using `np.arange()`.

Time both operations to see which one is faster!


```python
import numpy as np

L=range(10000000)
%timeit [i for i in L]
%time [i for i in L]
```

```python
a=np.arange(10000000)
%timeit a
%time a
```

# NumPy Temperature Converter : Convert an array of temperatures in Fahrenheit to Celsius using NumPy. Convert the temperatures to Kelvin as well.


```python
fahrenheit = np.array([32, 68, 100, 212])  
```

```python
celsius=(fahrenheit - 32) *5/9
kelvin = celsius + 273.15
```

```python
conversion_to_celsius=celsius
print(conversion_to_celsius)
```

```python
conversion_to_kelvin=conversion_to_celsius+273.15
print(conversion_to_kelvin)
```

# The Birthday Analyzer: Create a DataFrame with 10 people’s names and birthdays, then calculate their age based on today’s date.


```python
import pandas as pd


data = {
    'Name': ['John', 'Jane', 'Mike', 'Emily', 'David', 'Sarah', 'Michael', 'Olivia', 'Daniel', 'Sophia'],
    'Birthday': ['1990-01-01', '1992-05-15', '1985-12-20', '1995-08-10', '1988-03-25', '1998-11-03', '1980-07-14', '1993-09-28', '1982-04-12', '1997-06-28']
}
df = pd.DataFrame(data)
df
```

```python
from datetime import datetime

current_date = datetime.now().date()
df['Age']=current_date.year-df['Birthday'].apply(lambda x:datetime.strptime(x,'%Y-%m-%d').year)
df
```

# The Olympic Medal Counter: Load Olympic medal data and find:
The country with the most gold medals
The top 5 countries by total medals
The country with the highest medal-to-population ratio

```python
df = pd.read_csv("athlete_events.csv")

df.columns
```

```python
gold_medals_per_country = df.groupby('Team')['Medal'].count()
country_with_most_gold_medals = gold_medals_per_country.idxmax()
print(country_with_most_gold_medals)
```

```python


df_medals = df.dropna(subset=['Medal'])


total_medals_per_country = df_medals.groupby('Team')['Medal'].count()


top_5_countries = total_medals_per_country.sort_values(ascending=False).head(5)

print(top_5_countries)
```
